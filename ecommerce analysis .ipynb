# Converting date columns -datatype

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order year'] = df['Order Date'].dt.year
df['Order day of week'] = df['Order Date'].dt.dayofweek
```

```python
sale_by_month = df.groupby('Order month')['Sales'].sum().reset_index()
```

# analyze the sales and profit by customer segement

```python
df.head()
```

```python
sales_profit_by_segement = df.groupby('Segment').agg({'Sales':"sum",'Profit':'sum'}).reset_index()
```

```python
sales_profit_by_segement
```

```python
plt.figure(figsize=(17, 5))
sns.barplot(data =sales_profit_by_segement,x ='Segment',y ='Profit')
plt.show()

plt.figure(figsize=(17, 5))
sns.barplot(data =sales_profit_by_segement ,x ='Segment',y ='Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
segments =sales_profit_by_segement['Segment']
sales =sales_profit_by_segement['Sales']
profit =sales_profit_by_segement['Profit']

# Set bar width and positions
bar_width = 0.4
x = np.arange(len(segments))  # X locations for the groups

# Set figure size
plt.figure(figsize=(7, 3))

# Plot Sales bars
plt.bar(x - bar_width/2, sales, width=bar_width, label='Sales', color='blue', alpha=0.7)

# Plot Profit bars
plt.bar(x + bar_width/2, profit, width=bar_width, label='Profit', color='red', alpha=0.7)

# Labels and title
plt.xlabel("Segment")
plt.ylabel("Value")
plt.title("Sales and Profit by Segment")
plt.xticks(x, segments)  # Ensure segment names are correctly placed
plt.legend()  # Add legend

# Show plot
plt.show()
```

# Analyze the profit and sale ratio

```python
sales_profit_by_segement = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segement['Sales_to_Profit_ratio']=sales_profit_by_segement['Sales']/ sales_profit_by_segement['Profit']
print(sales_profit_by_segement[['Segment', 'Sales_to_Profit_ratio']])
```
